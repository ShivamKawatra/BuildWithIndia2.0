```python
# Display training results
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load the trained model
model = YOLO('runs/detect/train/weights/best.pt')

# Run validation to get final metrics
metrics = model.val()

print(f"ðŸ† Final Results:")
print(f"   mAP@0.5: {metrics.box.map50:.3f}")
print(f"   mAP@0.5-0.95: {metrics.box.map:.3f}")
print(f"   Precision: {metrics.box.mp:.3f}")
print(f"   Recall: {metrics.box.mr:.3f}")
```

```python
# Test model on sample images
import glob

# Get test images
test_images = glob.glob('test/images/*.png')[:5]

print("ðŸ” Testing model on sample images:")
for img_path in test_images:
    results = model(img_path)
    results[0].show()  # Display predictions
    
print("âœ… Model testing completed!")
```
